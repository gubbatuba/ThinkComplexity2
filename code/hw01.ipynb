#### Question 1: Exercise 2.2

Given that "reachable_nodes" is in O(n + m), "is_connected" can be said to also be in O(n + m) because the "reachable_nodes" is called once within it, and that would be the operation with dominance. The first operation simply selects a starting node, which is simply a constant time operation, as is the last check for list lengths equality. Thus, the O(n + m) operation wins out.

```python
def is_connected(G):
    start = next(G.nodes_iter())
    reachable = reachable_nodes(G, start)
    return len(reachable) == len(G)
```

```python
def all_pairs(nodes):
    for i, u in enumerate(nodes):
        for j, v in enumerate(nodes):
            if i < j:
                yield u, v
```

```python
def make_complete_graph(n):
    G = nx.Graph()
    nodes = range(n)
    G.add_nodes_from(nodes)
    G.add_edges_from(all_pairs(nodes))
    return G
```

#### Question 2: Exercise 2.3

```python
"""Original reachable_nodes function"""
def reachable_nodes(G, start):
    seen = set()
    stack = [start]
    while stack:
        node = stack.pop()
        if node not in seen:
            seen.add(node)
            stack.extend(G.neighbors(node))
    return seen
```

```python
"""'Optimized' reachable_nodes function"""
def reachable_nodes_2(G, start):
    seen = set()
    stack = [start]
    while stack:
        node = stack.pop()
        if node not in seen:
            seen.add(node)
            stack.extend(set(G.neighbors(node)) - seen) #only extend by those neighbors which have not been seen
    return seen
```

```python
complete = make_complete_graph(1000)
```

```python
%timeit len(reachable_nodes(complete,0))
```

```python
%timeit len(reachable_nodes_2(complete,0))
```

The second function is definitely more efficient, and it seems to make more of a difference the more edges / nodes that the complete graph contains. The second function performed 19 ms better than the first for a complete graph with 1000 nodes. This difference grows as the number of connections increases as well.

#### Question 3: Exercise 2.4

```python
def m_pairs(nodes, m):
    pairs = all_pairs(nodes)
    return sample(list(pairs), m)
```

```python
def make_m_graph(n,m):
    G = nx.Graph()
    nodes = range(n)
    G.add_nodes_from(nodes)
    G.add_edges_from(m_pairs(nodes,m))
    return G
```

```python
graph = make_m_graph(20,40)
nx.draw_circular(graph, 
                 node_color=COLORS[2], 
                 node_size=1000, 
                 with_labels=True)
```

```python
def prob_connected_2(n,m,iters=100):
    count = 0
    for i in range(iters):
        random_graph = make_m_graph(n,m)
        if is_connected(random_graph):
            count+=1
    return count/iters
```

```python
#looked at solution for this part
n = 10
pstar = np.log(n)/n
ps = np.logspace(-1.3, 0, 11)
ms = [int(p*n*(n-1)/2) for p in ps] #num of edges is based on p of edge between nodes
ys = [prob_connected_2(n,m) for m in ms]

for p, m, y in zip(ps,ms,ys):
    print(p,m,y)
```

```python
import thinkplot
thinkplot.vlines([pstar], 0, 1, color='gray')
thinkplot.plot(ps, ys)
thinkplot.config(xlabel='p', ylabel='prob connected', xscale='log', xlim=[ps[0], ps[-1]])
```

It seems that this ER graph has a probability of connectivity which increases and the transition is much smoother than that of the original graph. This ER graph shows a similar behavior to that of the original graph but for a lower number of nodes.
